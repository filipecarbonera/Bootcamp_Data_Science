###**Desafio 02:** Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Suldeste e Centro-Oeste) de cada estado (isso deu trabalho...).

```python
regioes = []
for estados in usaveis.index:
  if estados.startswith('1'):
    regioes.append('Norte')
  elif estados.startswith('2'):
    regioes.append('Nordeste')
  elif estados.startswith('3'):
    regioes.append('Sudeste')
  elif estados.startswith('4'):
    regioes.append('Sul')
  else:
    regioes.append('Centro-Oeste')

usaveis.insert(0, 'Regioes', regioes)

usaveis
```

```python
# 2. Tratar os dados populacionais:

dados_desafio_3_pop = dados_desafio_3_pop.drop([2,3,4,5,6,7,8], axis=0)
dados_desafio_3_pop = dados_desafio_3_pop.set_index('Unidade da Federacao')
dados_desafio_3_pop
```

```python
# 4. Criar a tabela per capta;

percapta = dados_desafio_3_anos / dados_desafio_3_pop
```
